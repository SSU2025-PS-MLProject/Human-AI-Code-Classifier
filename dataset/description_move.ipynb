```python
import pandas as pd
df = pd.read_csv('sample250_problem_list.csv')
df.head()
```

```python
import os
import shutil
```

```python
source_folder = r"C:/Users/user/주피터 파일/problem_descriptions"
target_folder = r"C:/Users/user/주피터 파일/target_folder"

# 타겟 폴더가 없으면 생성
os.makedirs(target_folder, exist_ok=True)

# id 리스트 추출
id_list = df['id'].astype(str).tolist()

# 폴더 내 파일 리스트
files_in_source = os.listdir(source_folder)

for file_name in files_in_source:
    # 파일이 .html이고, 이름(확장자 제외)이 id 리스트에 있으면 이동
    name, ext = os.path.splitext(file_name)
    if ext == '.html' and name in id_list:
        src_path = os.path.join(source_folder, file_name)
        dst_path = os.path.join(target_folder, file_name)
        shutil.move(src_path, dst_path)
        print(f"Moved: {file_name}")
```
