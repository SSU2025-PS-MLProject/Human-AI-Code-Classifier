```python
import numpy as np
import pandas as pd

df = pd.read_csv('problem_list.csv')
df.head()
```

```python
counts = df.value_counts(subset=['time_limit', 'memory_limit'])
counts_df = counts.reset_index(name='count')

counts_df
```

```python
sorted_counts = counts_df[counts_df['time_limit'] > 5000].sort_values(by='count', ascending=False)

print(sorted_counts)
```

```python
sorted_counts['count'].sum()
```

```python
top_3 = sorted_counts.head(3)[['time_limit', 'memory_limit']]
filtered_df = df.merge(top_3, on=['time_limit', 'memory_limit'])
```

```python
filtered_df.shape
```

```python
filtered_df.head()
```

```python
sampled_df = filtered_df.sample(n=250, random_state=100)

sampled_df.shape
```

```python
sampled_df.head()
```

```python
sampled_df = sampled_df.sort_values(by='id')
sampled_df.to_csv('sample250_problem_list.csv', index=False)
```
